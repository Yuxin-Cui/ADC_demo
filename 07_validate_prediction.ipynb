```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

np.random.seed(42)

############################################
# RETROSPECTIVE VALIDATION
############################################

# -------------------------------
# Performance
# -------------------------------
observed_synergy = np.random.normal(0.3, 0.15, 200)
predicted_synergy = observed_synergy + np.random.normal(0, 0.08, 200)

r, _ = pearsonr(observed_synergy, predicted_synergy)

plt.figure(figsize=(4, 3))
plt.scatter(observed_synergy, predicted_synergy, s=12)
plt.xlabel("Observed AUC")
plt.ylabel("Predicted AUC")
plt.ylim(0, 1)
plt.title(f"Performance (r = {r:.2f})")
plt.tight_layout()
plt.show()


# -------------------------------
# Robustness
# -------------------------------
train_pred = np.random.normal(0.4, 0.1, 300)
train_obs  = train_pred + np.random.normal(0, 0.05, 300)

hidden_pred = np.random.normal(0.4, 0.1, 120)
hidden_obs  = hidden_pred + np.random.normal(0, 0.10, 120)

plt.figure(figsize=(4, 3))
plt.scatter(train_obs, train_pred, label="Train", s=12)
plt.scatter(hidden_obs, hidden_pred, label="Hidden", s=12)
plt.xlabel("Observed AUC")
plt.ylabel("Predicted AUC")
plt.ylim(0, 1)
plt.legend(fontsize=8)
plt.title("Robustness")
plt.tight_layout()
plt.show()


# -------------------------------
# LOPO-CV (LOPO-CV)
# -------------------------------
pairs = ["AB", "AC", "BD", "CD", "EF"]
lopo_scores = np.random.normal(0.65, 0.05, len(pairs))

plt.figure(figsize=(4, 3))
plt.bar(pairs, lopo_scores)
plt.axhline(np.mean(lopo_scores), linestyle="--", linewidth=1)
plt.ylabel("Correlation")
plt.ylim(0, 1)
plt.title("LOPO-CV")
plt.tight_layout()
plt.show()


# -------------------------------
# Temporal Back-testing
# -------------------------------
years = ["2023", "2024", "2025"]
temporal_perf = [0.72, 0.68, 0.66]

plt.figure(figsize=(4, 3))
plt.plot(years, temporal_perf, marker="o")
plt.ylabel("Correlation")
plt.ylim(0, 1)
plt.title("Temporal")
plt.tight_layout()
plt.show()


############################################
# PROSPECTIVE VALIDATION
############################################

# -------------------------------
# Selection
# -------------------------------
groups = ["Top", "Mid", "Random"]
hit_rates = [0.45, 0.22, 0.08]

plt.figure(figsize=(4, 3))
plt.bar(groups, hit_rates)
plt.ylabel("Hit Rate")
plt.ylim(0, 1)
plt.title("Selection")
plt.tight_layout()
plt.show()


# -------------------------------
# Dependency
# -------------------------------
conditions = ["A+/B+", "A+/B", "A/B+"]
killing = [85, 35, 30]

plt.figure(figsize=(4, 3))
plt.bar(conditions, killing)
plt.ylabel("Killing (%)")
plt.ylim(0, 100)
plt.title("Dependency")
plt.tight_layout()
plt.show()


# -------------------------------
# Benchmarking
# -------------------------------
constructs = ["Bispecific", "Mono A", "Mono B", "Random"]
efficacy = [90, 55, 50, 40]

plt.figure(figsize=(4, 3))
plt.bar(constructs, efficacy)
plt.ylabel("Max Killing (%)")
plt.ylim(0, 100)
plt.title("Benchmarking")
plt.tight_layout()
plt.show()


# -------------------------------
# Enrichment
# -------------------------------
methods = ["Random", "Prediction-guided"]
final_hits = [0.07, 0.42]

plt.figure(figsize=(4, 3))
plt.bar(methods, final_hits)
plt.ylabel("Validated Hit Rate")
plt.ylim(0, 1)
plt.title("Enrichment")
plt.tight_layout()
plt.show()
```
