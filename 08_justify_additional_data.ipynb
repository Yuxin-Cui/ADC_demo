```python
import matplotlib.pyplot as plt
import numpy as np

# Inclusion criteria
criteria = ['Predictive Accuracy', 'Biological Realism', 'Translational Relevance', 'Wet-Lab Risk Reduction']
num_criteria = len(criteria)

# High Priority data scores (more varied demo values)
high_priority = {
    'Monospecific ADC':      [0.95, 0.65, 0.55, 0.85],
    'Target expression':     [0.75, 0.90, 0.70, 0.60],
    'Proteomics':            [0.60, 0.80, 0.50, 0.65],
    'Structural info':       [0.85, 0.60, 0.65, 0.70],
    'CRISPR / functional':   [0.80, 0.75, 0.80, 0.90]
}

# Medium Priority data scores (more varied demo values)
medium_priority = {
    'PPI networks':          [0.50, 0.65, 0.40, 0.45],
    'PDX / organoids':       [0.45, 0.85, 0.95, 0.55],
    'Spatial transcript':    [0.60, 0.90, 0.75, 0.65],
    'PTMs / isoforms':       [0.35, 0.70, 0.55, 0.50]
}

def plot_radar(data_dict, title):
    # Set up radar plot
    angles = np.linspace(0, 2 * np.pi, num_criteria, endpoint=False).tolist()
    angles += angles[:1]  # close the loop
    
    plt.figure(figsize=(5,5))
    ax = plt.subplot(111, polar=True)
    
    for name, scores in data_dict.items():
        values = scores + scores[:1]  # close the loop
        ax.plot(angles, values, label=name, linewidth=1.5)
        ax.fill(angles, values, alpha=0.1)
    
    # Customize fonts to black
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(criteria, color='black', fontsize=10)
    ax.tick_params(axis='y', colors='black', labelsize=9)
    ax.set_ylim(0, 1)
    plt.title(title, size=14, y=1.1, color='black')
    plt.legend(fontsize=8, bbox_to_anchor=(1.05, 1), facecolor='white')
    plt.tight_layout()
    plt.show()

# Plot High Priority
plot_radar(high_priority, "High Priority Data")

# Plot Medium Priority
plot_radar(medium_priority, "Medium Priority Data")
```
